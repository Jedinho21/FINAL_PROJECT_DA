# Which student spends the most ?

```python
df1['total_expense'] = df1[['housing', 'food', 'transportation', 'books_supplies', 'entertainment', 'personal_care', 'technology', 'health_wellness', 'miscellaneous']].sum(axis=1)
```

```python
max_spender = df1.loc[df1['total_expense'].idxmax()]
```

```python
print("The student who spends the most is:")
print(max_spender[['age', 'gender', 'year_in_school', 'major', 'total_expense','preferred_payment_method']])
```

```python
import matplotlib.pyplot as plt

# Assuming df1 is your DataFrame containing student data
# Assuming max_spender is the row containing information about the student who spends the most

# Extracting relevant information
categories = ['housing', 'food', 'transportation', 'books_supplies', 'entertainment', 'personal_care', 'technology', 'health_wellness', 'miscellaneous']
expenses = max_spender[categories]

# Plotting the expenses
plt.figure(figsize=(10, 6))
plt.bar(categories, expenses, color='skyblue')
plt.title('Expenses Breakdown for the Student Who Spends the Most')
plt.xlabel('Expense Categories')
plt.ylabel('Amount ($)')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.tight_layout()  # Adjust layout to prevent clipping of labels
plt.show()
```
